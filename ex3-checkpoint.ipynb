```python
# Import all necessary libraries.
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=150, n_features=2, centers=5, cluster_std=1.0, shuffle=True, random_state=42)
```

```python
print(X)
```

```python
plt.scatter(X,y)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('a) figura com os dados de treinamento')
plt.show()
```

```python
X[0] = [1,4]
X[1] = [4,3]
X[2] = [4,5]
X[3] = [3,6]
X[4] = [6,7]
X[5] = [3,3]
X[6] = [2,5]
X[7] = [2,2]
X[8] = [2,3]

centroids[0] = [5,3]
centroids[1] = [1,3]
centroids[2] = [3,4]
```

```python
x1 = [1,4,4,3,6,3,2,2,2]
x2 = [4,3,5,6,7,3,5,2,3]

plt.scatter(x1,x2)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.show()
```

```python
km = KMeans(n_clusters=3, init=centroids)

km.fit(X)
```

```python
print('valores óptimos para centroids')
for i in range(0,3):
    print('Centroid %d: (%1.2f, %1.2f)' % (i,km.cluster_centers_[i][0],km.cluster_centers_[i][1]))
```

```python
print('Foram necessárias',km.n_iter_, 'Iterações para treinar o modelo')
```

```python
y = km.labels_
```

```python
idx0 = np.argwhere(y == 0)
idx1 = np.argwhere(y == 1)
idx2 = np.argwhere(y == 2)

fig = plt.figure(figsize=(5,5))
plt.plot(X[idx0,0], X[idx0,1], 'rx', color='blue', label='Class 0')
plt.plot(X[idx1,0], X[idx1,1], 'rx', color='red', label='Class 1')
plt.plot(X[idx2,0], X[idx2,1], 'rx', color='violet', label='Class 2')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('e) clusteres a que cada um pertence')
plt.legend()
plt.show()
```
